```python
DUMMY_SEP = '__'
```

```python
adult_train_dummies = pd.get_dummies(adult_train, prefix_sep=DUMMY_SEP)
```

```python
adult_train_dummies
```

```python
train_dummy_columns = [c for c in adult_train_dummies.columns if DUMMY_SEP in c]
```

```python
dummy_columns_to_remove = [c for c in train_dummy_columns if adult_train_dummies[c].sum() < 10] 
```

```python
dummy_columns_to_remove
```

```python
adult_train_dummies = adult_train_dummies.drop(dummy_columns_to_remove, axis=1)
```

```python
adult_train_dummies
```

```python
def get_dummies_for_test(train_columns, test_df, dummy_sep):
    initial_df = pd.get_dummies(test_df, prefix_sep=dummy_sep)
#     missing_in_train = set(test_df.columns) - set(train_columns)
    missing_in_test = set(train_columns) - set(initial_df.columns)
    
    for c in missing_in_test:
        initial_df[c] = 0
    return initial_df.loc[:,train_columns].copy()
```

```python
adult_test_dummies = get_dummies_for_test(adult_train_dummies.columns, adult_test, DUMMY_SEP)
adult_val_dummies = get_dummies_for_test(adult_train_dummies.columns, adult_val, DUMMY_SEP)
```

```python
adult_val_dummies
```

```python
X_train, y_train = adult_train_dummies.drop("income >50K",axis=1), adult_train_dummies['income >50K']
X_test, y_test = adult_test_dummies.drop("income >50K",axis=1), adult_test_dummies['income >50K']
X_val, y_val = adult_val_dummies.drop("income >50K",axis=1), adult_val_dummies['income >50K']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
```

```python
dtree_clf = DecisionTreeClassifier(min_samples_split=20)
```

```python
dtree_clf.fit(X_train, y_train)
```

```python
accuracy_score(y_val, dtree_clf.predict(X_val))
```

```python
1 - y_val.sum() / len(y_val)
```

```python
print(classification_report(y_train, dtree_clf.predict(X_train)))
print(classification_report(y_val, dtree_clf.predict(X_val)))
```

```python
probs = dtree_clf.predict_proba(X_val)[:,1]
```

```python
fpr, tpr, thresholds = roc_curve(y_val, probs)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot([0,1],[0,1])
plt.show()
```

```python
roc_auc_score(y_val, probs)
```

```python
logr_cls = LogisticRegression()
```

```python
logr_cls.fit(X_train, y_train)
```

```python
print(classification_report(y_train, logr_cls.predict(X_train)))
print(classification_report(y_val, logr_cls.predict(X_val)))
```

```python
grad_boosting = GradientBoostingClassifier()
```

```python
grad_boosting.fit(X_train, y_train)
```

```python
print(classification_report(y_train, grad_boosting.predict(X_train)))
print(classification_report(y_val, grad_boosting.predict(X_val)))
```

```python
fprb, tprb, _ = roc_curve(y_val, grad_boosting.predict_proba(X_val)[:,1])
fprl, tprl, _ = roc_curve(y_val, logr_cls.predict_proba(X_val)[:,1])
fprt, tprt, _ = roc_curve(y_val, dtree_clf.predict_proba(X_val)[:,1])

plt.plot(fprt, tprt, label='decision_tree')
plt.plot(fprb, tprb, label='grad_boosting')
plt.plot(fprl, tprl, label='logistic regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot([0,1],[0,1])
plt.legend()
plt.show()
```

```python
roc_auc_score(y_val, grad_boosting.predict_proba(X_val)[:,1])
```
